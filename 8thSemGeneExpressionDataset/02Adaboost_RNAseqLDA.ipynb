# Implementation of Gene Classification

1. Using Adaboost as a classifier
2. Using Linear Discriminant Analysis as Dimensionality Reduction
3. Using KFold Cross Validation
4. Encoding the categorical labels: LabelEncoder

```python
import pandas as pd
import numpy as np
import time as t
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
#loading the data as a pandas dataframe
data = pd.read_csv('RNAseq.gz', compression='gzip')
```

```python
#extracting target labels
labels = data[['gene_0']]
```

```python
#preparing the data
#Remove column 'gene_0' as it contain target labels
df1 = data.drop('gene_0', axis=1)
```

```python
df = df1.iloc[:801]
#df now contains only the features columns
```

```python
df.head()
```

```python
df.tail()
```

```python
#Remove the column 'TCGA-PANCAN-HiSeq-801x20531/' as it only contains the sample number and it is not needed for training
df=df.drop('TCGA-PANCAN-HiSeq-801x20531/', axis=1)
```

```python
df.head()
```

```python
#instantiate StandardScaler
scaler = StandardScaler()
```

```python
#scaling the data
scaler.fit(df)
```

```python
scaled_features = scaler.transform(df)
```

```python
#Encoding the categorical target labels
le = preprocessing.LabelEncoder()
le.fit(labels)
le.classes_
```

```python
y=le.transform(labels)
```

```python
y.shape
```

```python
adaboost = AdaBoostClassifier(n_estimators=100, random_state = 1)
```

```python
#finding the optimal n_components (optimal number of components that we need) by comparing the accuracy for each number of components

results = []

for i in range(1,5):
    fs = LDA(n_components=i).fit(scaled_features, y)
    dfnew = fs.fit_transform(scaled_features, y)
    kfold = KFold(n_splits=10)
    results.append(cross_val_score(adaboost, dfnew, y, cv=kfold))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,5),results,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Number of components vs. Accuracy')
plt.xlabel('Number of components')
plt.ylabel('Accuracy')
```

```python
#Now that we take n_components=2 (based on the graph) we can extract the best 2 features
featureSelection = LDA(n_components=2).fit(scaled_features,y)
SelectedFeatures = featureSelection.fit_transform(scaled_features, y)
```

```python
#check the hyperparameters of the model
print(adaboost.get_params())
```

```python
#Tuning the model (first with RandomizedSearchCV)

random_grid = {'n_estimators': [50, 100, 200, 400], 'learning_rate': [0.001,0.01, 0.1,0.2,0.5]}
print(random_grid)
```

```python
X_train, X_test, y_train, y_test = train_test_split(SelectedFeatures, y, test_size=0.25, random_state=42)
```

```python
# Use the RandomizedSearchCV and GridSearchCV to search for best hyperparameters
# First create the base model to tune
ada = AdaBoostClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
ada_random = RandomizedSearchCV(estimator = ada, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
ada_random.fit(X_train, y_train)
```

```python
ada_random.best_params_
```

```python
#function to calculate the accuracy
def evaluate(model, test_features, test_labels):
    y_pred = model.predict(test_features)
    return accuracy_score(y_test, y_pred)
```

```python
#testing the base model ie without the tuned hyperparameters
base_model = AdaBoostClassifier(random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
print(base_accuracy)
```

```python
#testing the model with the tuned hyperparameters
best_random = ada_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)
print(random_accuracy)
```

```python
print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```

```python
#Tuning model (now using more focused values that we got from RandomizedSearchCV to GridSearchCV)
# Create the parameter grid based on the results of random search 
param_grid = {'n_estimators': [190, 200,250],
              'learning_rate': [0.01, 0.009,0.015,0.02]
    }

ada = AdaBoostClassifier()

grid_search = GridSearchCV(estimator = ada, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_
grid_accuracy = evaluate(best_grid, X_test, y_test)
print(grid_accuracy)
```

```python
print('Improvement of {:0.2f}%.'.format( 100 * (grid_accuracy - base_accuracy) / base_accuracy))
```
